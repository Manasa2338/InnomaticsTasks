**Problem Statement 1:** Smart Parking Lot Management System

```python
def smart_parking_system():
    capacity = int(input("Enter Parking Capacity: "))

    n = int(input("Enter number of vehicle logs: "))
    vehicle_logs = []

    print("Enter vehicle logs (IN/OUT):")
    for _ in range(n):
        log = input().strip().upper()
        vehicle_logs.append(log)

    current_parked = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1

        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    print("\nCurrently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)

    if current_parked > capacity:
        print("Parking Status: Exceeded Capacity ðŸš¨")
    else:
        print("Parking Status: Available")


smart_parking_system()
```

**Problem Statement 2:** Online Food Delivery Time Estimator


```python
def estimate_delivery_time():
    distance = float(input("Enter Distance (km): "))
    traffic = input("Enter Traffic Level (Low/Medium/High): ").strip().lower()
    weather = input("Enter Weather (Clear/Rainy/Stormy): ").strip().lower()

    # Base speed assumption (km per hour)
    base_speed = 30

    # Base delivery time in minutes
    eta = (distance / base_speed) * 60

    # Traffic adjustments
    if traffic == "medium":
        eta += 10
    elif traffic == "high":
        eta += 20

    # Weather adjustments
    if weather == "rainy":
        eta += 15
    elif weather == "stormy":
        eta += 25

    print("\nEstimated Delivery Time:", round(eta), "minutes")


estimate_delivery_time()
```

**Problem Statement 3:** Movie Theatre Seat Occupancy Analyzer

```python
def analyze_seat_booking():
    total_seats = int(input("Enter Total Seats: "))
    n = int(input("Enter number of booked seat entries: "))

    booked_seats = []
    print("Enter booked seat entries (1 for booked):")
    for _ in range(n):
        seat = int(input())
        booked_seats.append(seat)

    total_booked = sum(booked_seats)

    occupancy = (total_booked / total_seats) * 100

    print("\nOccupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider Opening Additional Show")
    else:
        print("Show Status: Seats Available")


analyze_seat_booking()
```

**Problem Statement 4:** Cloud Server Load Classification System

```python
def classify_server_load():
    n = int(input("Enter number of CPU readings: "))

    readings = []
    print("Enter CPU readings (%):")
    for _ in range(n):
        value = float(input())
        readings.append(value)

    total = 0
    for cpu in readings:
        total += cpu

    average = total / n

    print("\nAverage CPU Load:", round(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


classify_server_load()
```

 **Problem Statement 5:** Smart Classroom Resource Usage Monitor


```python
def track_resource_usage():
    n = int(input("Enter number of resources: "))

    usage = {}

    print("Enter Resource Name and Usage Hours:")
    for _ in range(n):
        name = input("Resource Name: ")
        hours = float(input("Usage Hours: "))
        usage[name] = hours

    # Define overuse threshold (example: > 8 hours)
    threshold = 8

    overused = []

    for resource, hours in usage.items():
        if hours > threshold:
            overused.append(resource)

    if overused:
        print("\nOverused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("\nOverused Resources: None")
        print("Energy Alert: No")


track_resource_usage()
```

**Problem Statement 6:** Online Event Registration Capacity Controller

```python
def manage_event_registration():
    capacity = int(input("Enter Event Capacity: "))
    total_registrations = int(input("Enter Total Registrations: "))

    confirmed = 0
    waitlisted = 0

    # Process registrations one by one
    for i in range(1, total_registrations + 1):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("\nConfirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


manage_event_registration()
```
