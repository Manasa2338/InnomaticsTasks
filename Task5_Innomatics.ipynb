**Problem 1:** Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    # Initialize total likes counter
    total_likes = 0

    # Loop through the list to calculate total likes
    for likes in likes_list:
        total_likes += likes

    # Determine post status based on total likes
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Display results in required format
    print("Total Likes:", total_likes)
    print("Post Status:", status)
likes_data = [200, 300, 150, 400]   # Sample likes from different sources
analyze_engagement(likes_data)
```

**Problem 3:** Agriculture – Rainfall Adequacy Checker

```python
# Function to analyze rainfall data
def analyze_rainfall(rainfall_list, required_level):
    total = 0  # store total rainfall

    # Calculate total rainfall
    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)  # calculate average

    # Check rainfall status
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)


# User input
n = int(input("Enter number of rainfall readings: "))
rainfall = []

for i in range(n):
    value = int(input("Enter rainfall value: "))
    rainfall.append(value)

required = int(input("Enter required rainfall level: "))

# Call function
analyze_rainfall(rainfall, required)
```

**Problem 4:** Social Media – Duplicate Account Detection

```python
# Function to check duplicate usernames
def check_duplicates(usernames):
    unique_users = set(usernames)  # convert list to set

    # If length differs, duplicates exist
    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# User input
n = int(input("Enter number of usernames: "))
users = []

for i in range(n):
    name = input("Enter username: ")
    users.append(name)

# Call function
check_duplicates(users)
```

**Problem 6:** Agriculture – Premium Crop Price Filter

```python
# Function to filter premium crops
def filter_premium_crops(prices):

    premium = []  # list to store premium crop prices

    # Filter prices greater than 2000
    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


# Take input from user
n = int(input("Enter number of crop prices: "))
prices = []

for i in range(n):
    value = int(input("Enter crop price: "))
    prices.append(value)

# Call function
filter_premium_crops(prices)
```

**Problem 7:** System Monitoring – Application Health Checker


```python
# Function to evaluate application health
def check_application_health(errors):

    # Check error conditions
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    # Print result in required format
    print("Error Count:", errors)
    print("System Status:", status)


# Take input from user
error_count = int(input("Enter error count: "))

# Call function
check_application_health(error_count)
```

**Problem 8:** Banking – Daily Transaction Limit Checker

```python
# Function to check daily transaction limit
def check_transaction(amount):

    limit = 50000  # daily transaction limit

    # Check if amount is within limit
    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    # Print result in required format
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Take input from user
amount = int(input("Enter transaction amount: "))

# Call function
check_transaction(amount)
```

**Problem 9:** E-Learning – Student Attendance Eligibility System


```python
# Function to calculate attendance and check eligibility
def check_attendance(attendance_list):

    total_classes = len(attendance_list)  # total number of classes
    attended = 0  # count attended classes

    # Count present days (1 means present)
    for day in attendance_list:
        if day == 1:
            attended += 1

    percentage = (attended / total_classes) * 100  # calculate percentage

    # Check eligibility
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


# User input
n = int(input("Enter total number of classes: "))
attendance = []

print("Enter 1 for Present and 0 for Absent")

for i in range(n):
    value = int(input("Enter attendance: "))
    attendance.append(value)

# Call function
check_attendance(attendance)
```

**Problem 10:** Smart Electricity Bill Analyzer

```python
# Function to calculate electricity bill
def calculate_bill(units):

    # Calculate bill based on slab rates
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Classify usage level
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# User input
units = int(input("Enter units consumed: "))

# Call function
total_bill, usage_status = calculate_bill(units)

print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
