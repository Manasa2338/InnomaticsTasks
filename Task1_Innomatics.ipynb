**1. User Login Check**


```python
# Stored credentials
stored_username = "admin"
stored_password = "1234"

# Given credentials (can be replaced with input())
username = "admin"
password = "1234"

# Check login
if username == stored_username and password == stored_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

**2. Pass / Fail Analyzer**

```python
# List of student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Loop through each student's marks
for mark in marks:
    if mark >= 50:
        # Student has passed
        pass_count += 1
    else:
        # Student has failed
        fail_count += 1

# Display final results
print("Final Result--")
print(f"Total Students : {len(marks)}")
print(f"Passed Students: {pass_count}")
print(f"Failed Students: {fail_count}")
```

**3. Simple Data Cleaner**

```python
# Original list of names
names = [" Alice ", "bob", " CHARLIE "]

# List to store cleaned names
cleaned_names = []

# Process each name
for name in names:
    # Remove extra spaces using strip()
    name = name.strip()

    # Convert to lowercase
    name = name.lower()

    # Add cleaned name to new list
    cleaned_names.append(name)

# Display cleaned list
print("Cleaned Names:", cleaned_names)
```

**4. Message Length Analyzer**

```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Loop through each message
for message in messages:

    # Calculate length of the message
    length = len(message)

    # Print message and its length
    print(f"Message: \"{message}\" | Length: {length}")

    # Check if message is longer than 10 characters
    if length > 10:
        print("→ This message is longer than 10 characters.")

    print()
```

**5. Error Message Detector**

```python
# List of system log entries
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initialize error counter
error_count = 0

# Loop through each log entry
for log in logs:
    if log == "ERROR":
        error_count += 1

# Display total error count
print("Total ERROR entries:", error_count)
```
